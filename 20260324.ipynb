```python
# 도전미션
리스트 numbers 각 원소는 숫자 0부터 9까지로 이루어져 있습니다. 이때, numbers에서 연속적으로 나타나는 숫자는 하나만 남기고 전부 제거하려고 합니다. 단, 제거된 후 남은 수들은 numbers의 원소의 순서를 유지해야 합니다. 

예를 들면,
numbers = [1, 1, 3, 3, 0, 1, 1] ->  [1, 3, 0, 1] 
numbers = [4, 4, 4, 3, 3] ->  [4, 3]

코드를 작성해 주세요
```

```python
numbers = [4, 4, 4, 3, 3]

answer = []
comp = numbers[0]
answer.append(comp)

for i in numbers:
    if comp != i:
        comp = i
        answer.append(comp)
answer
```

```python
def max_num(a, b, c):
    max_val = a
    if b > max_val:
        max_val = b
    if c > max_val:
        max_val = c

    return max_val
```

```python
num = max_num(-3, -5, -1)
num
```

```python
def num_min(arr):
    # print(arr)
    min_val = arr[0]
    for i in arr:
        if min_val > i:
            min_val = i

    return min_val
```

```python
num_min([-3, -5, -1])
```

```python
num_min([52, 273, 32 ,103, 275, 24, 57]) #24
```

```python
def add_many(*args):
    print(args)
    # 입력 정수들의 합을 리턴
    result = 0
    for i in args:
        result += i
    return result
```

```python
add_many(1,2,3) # 6
```

```python
add_many(1, 2, 3, 4, 5) # 15
```

```python
def add_mul(selection ,*args):
    print(selection, args)
    # 입력 정수들의 합을 리턴
    if selection == 'add':
        result = 0
        for i in args:
            result += i
    elif selection == 'mul':
        result = 1
        for i in args:
            result *= i
    else:
        print('연산자 종류 확인할것!!!')
        result = None
    return result
```

```python
add_mul('add', 1,2,3)
```

```python
add_mul('mul', 1,2,3,4,5)
```

```python
add_mul('sub', 1,2,3,4,5)
```
