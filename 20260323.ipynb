```python
eng = 90
math = 100

if eng >= 90 and math >= 90:
    print('스티커')
else:
    print('좀 더 노력하세요')
```

```python
둘다 90점 이상이면 "둘 다 성공"
영어는 90점이상인데 수학은 그렇지 않으면 "좀 더 노력"
둘다 90점 미만이면 "많이 노력"
```

```python
eng = 90
math = 100

if eng >= 90:
    if math >= 90:
        print("둘 다 성공")
    else:
        print("좀 더 노력")
else:
    print('많이 노력')
```

```python
# 미션: 성적처리프로그램
90점 이상 100점 미만 A 학점
80점 이상 90점 미만 B 학점
70점 이상 80점 미만 C 학점
60점 이상 70점 미만 D 학점
60점 미만 F 학점
```

```python
score = 69 # int(input("점수를 입력하세요: "))

if score >= 90:
    print('A')
elif score >= 80:
    print('B')
elif score >= 70:
    print('C')
elif score >= 60:
    print('D')
else:
    print('F')
```

```python
# 1부터 10까지의 합을 구하되, 합이 20 이상이 되는 때의 정수를 구하시오. 
hap = 0

for i in range(1, 10 + 1):
    # print(i)
    hap +=  i
    if hap >= 20:
        print(f"정수: {i}, 합계: {hap}")
        break
hap
```

```python
# 숫자맞추기 게임 1~100 / 몇번만에 맞췄는지...도 출력

import random

answer = random.randint(1, 100)
print(answer)

while True:
    # 사용자입력
    guess = int(input("1부터 100사의 숫자를 맞춰주세요: "))
    if guess == answer:
        print('정답입니다')
        break
    elif guess > answer:
        print('너무 커요')
    else:
        print('너무 커요')

print('게임종료')
```
